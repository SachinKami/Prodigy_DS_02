```python
import pandas as pd

# Load the dataset
data = pd.read_csv('C:/Users/Sachin Kamath/Downloads/Titanic_Dataset.csv')
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)
```

```python
#replace missing values in the "Age" column with the mean
data['Age'].fillna(data['Age'].mean(), inplace=True)
```

```python
#replace missing values in the "Embarked" column with the mode:
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of passenger ages
plt.figure(figsize=(8, 5))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Explore the survival rate by class
sns.countplot(data['Pclass'], hue=data['Survived'])
plt.title('Survival Count by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()

# Explore the survival rate by gender
sns.countplot(data['Sex'], hue=data['Survived'])
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```
